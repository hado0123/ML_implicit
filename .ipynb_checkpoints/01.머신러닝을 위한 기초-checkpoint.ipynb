### 3. 데이터 프레임을 활용해 housing_median_age가 10 미만인 데이터 삭제

```python

train = pd.read_csv('sample_data/california_housing_train.csv')

new_data = []

# iterrows(): 데이터 프레임의 각 행을 반복(iterate) 하도록 해줌 
# index: 현재 행의 인덱스 (정수형 또는 데이터프레임에서 사용된 인덱스 값).
# row: 현재 행의 데이터를 판다스 Series 형태로 반환.

for index, row in train.iterrows(): 
  # print(index)
  if row.housing_median_age >= 10: # housing_median_age가 10 이상인 데이터만 
    new_data.append(row)

new_data
```

```python
new_train = pd.DataFrame(new_data) # 다시 행열이 있는 DataFrame형태도 변경
new_train
```

### 4. 특정 데이터 변경

```python
train = pd.read_csv('sample_data/california_housing_train.csv')
print(train.loc[58]) # 58번째 행 조회
```

```python
# loc[row_label, column_label] 형태로, 행과 열을 동시에 선택할 수 있음

# train['housing_median_age'] < 10
# housing_median_age 값이 10보다 작은 모든 행을 필터링

# train.loc[condition, 'housing_median_age']
# 조건을 만족하는 행의 housing_median_age 열만 선택

# = train.loc[...] * 10
# 선택된 값들을 10배로 수정

train.loc[train['housing_median_age'] < 10, 'housing_median_age'] = train.loc[train['housing_median_age'] < 10, 'housing_median_age'] * 10

print(train.loc[58]) # 변경된 값 다시 조회
```

### 6. pyplot을 사용한 그래프 생성

```python
import matplotlib.pyplot as plt # Python에서 데이터를 시각화하기 위한 가장 기본적인 라이브러리. plt로 줄여서 사용.
import seaborn as sns # 고급 그래프를 그릴 수 있는 시각화 라이브러리. matplotlib 위에서 동작하며, 더 세련된 스타일과 옵션 제공.
```

```python
train = pd.read_csv('sample_data/california_housing_train.csv')

# 선택된 열을 히스토그램 형태로 플롯. 히스토그램은 데이터를 구간별로 나눠 각 구간의 빈도를 나타내는 그래프.
# 히스토그램은 분포를 이해하는 데 유횽
train['housing_median_age'].plot(kind='hist') 
```

```python
# Box Plot은 데이터의 분포를 요약하고 이상값(outlier)을 시각적으로 탐지하는 데 유용한 도구

# 중앙값(Median): 약 30
# 중위 50% 데이터(Q1 ~ Q3): 약 20 ~ 40 사이.
"""
상자의 아래 경계(Q1, 1사분위수):데이터의 하위 25%에 해당하는 값의 경계
하위 25%는 약 20

상자의 중앙선(중앙값, Median): 데이터의 중앙값(50번째 백분위수)
중앙값은 약 30

상자의 위 경계(Q3, 3사분위수): 데이터의 상위 25%를 제외한 값의 경계
상위 25%는 약 40
"""

# 최소값과 최대값: 0, 50
# 이상값: 없음.

train['housing_median_age'].plot(kind='box')
```

```python
"""
pd.cut: 데이터를 특정 구간(bin)으로 나눕니다.
매개변수:
 - train['housing_median_age']: 나눌 데이터 열입니다.
 - 3: 데이터를 3개의 구간으로 나눕니다.(구간은 동일한 크기로 나뉩니다. 첫 번째 구간: 0~16.67, 두 번째 구간: 16.67~33.33, 세 번째 구간: 33.33~50)
 - labels=["new", "medium", "old"]: 각 구간에 대해 "new", "medium", "old"이라는 라벨을 부여합니다.
"""

train['housing_median_age_bin'] = pd.cut(train['housing_median_age'], 3, labels=["new", "medium", "old"])

# train['housing_median_age_bin'].value_counts(): 각 범위(라벨)에 해당하는 데이터 개수를 세는 함수입니다.
"""
new: 약 4500개
old: 약 5100개
medium: 약 7200개

정도임을 알 수 있습니다.
"""
train['housing_median_age_bin'].value_counts().plot(kind='bar')

```

```python
train['housing_median_age_bin'] = pd.cut(
    train['housing_median_age'], 
    bins=[0, 15, 30, 50],  # 사용자 정의 구간(첫 번째 구간: 0~15, 두 번째 구간: 16~30, 세 번째 구간: 31~50)
    labels=["new", "medium", "old"]
)


# 그래프 꾸미기
plt.title("Housing Median Age Distribution")
plt.xlabel("Age Category")
plt.ylabel("Count")
train['housing_median_age_bin'].value_counts().plot(kind='bar', color='skyblue')
plt.show()
```

```python
#  total_rooms 열과 median_house_value 열의 데이터를 이용해 산점도(Scatter Plot)를 그립니다. 산점도는 두 변수 간의 관계(상관관계)를 시각화하는 데 사용됩니다.
# total_rooms: 전체 방수, median_house_value: 주택 중간 가격
train.plot(x='total_rooms', y='median_house_value', kind='scatter')
```

### 7. 사이킷런(Scikit-learn) 사용하기
- 머신러닝과 데이터 분석을 쉽게 구현할 수 있도록 지원하는 오픈 소스 라이브러리

#### 다양한 머신러닝 알고리즘 제공
- 지도 학습 (Supervised Learning):
    - 회귀(Regression): 선형 회귀, 리지 회귀, 라쏘 회귀 등.
    - 분류(Classification): 로지스틱 회귀, 의사결정트리, 랜덤 포레스트, SVM, K-최근접 이웃(KNN) 등.
- 비지도 학습 (Unsupervised Learning):
    - 클러스터링(Clustering): K-평균(K-means), DBSCAN, 계층적 클러스터링 등.
    - 차원 축소(Dimensionality Reduction): PCA, t-SNE 등.
- 강화 학습 (Reinforcement Learning):
    - 사이킷런은 기본적으로 강화 학습을 포함하지 않음(별도 라이브러리 사용).

#### 1. LabelEncoder 사용하기

```python
from sklearn.preprocessing import LabelEncoder # 사이킷런 LabelEncoder 패키지

# LabelEncoder는 문자 라벨(label)을 정수형으로 변환시킬때 사용됩니다.(알파벳 순서 기준)
```

```python
# 기본 개념
# 범주형 데이터(문자 데이터)를 숫자 데이터로 변환하는 예제
a = ['a', 'b', 'c']

"""
LabelEncoder 객체 생성. 
이 객체는 문자형 라벨을 숫자형으로 변환하거나, 반대로 숫자형을 문자형으로 복원할 때 사용됩니다.
"""
le = LabelEncoder()

"""
fit(a): 데이터 a의 고유값(라벨)을 학습합니다.
학습된 라벨: 'a', 'b', 'c'
정수형 값으로 매핑: 
'a' → 0
'b' → 1
'c' → 2
"""
le.fit(a)

"""
학습된 라벨을 기준으로 데이터를 숫자로 변환합니다. 변환결과: ['a', 'b', 'c'] → [0, 1, 2]
"""
le.transform(a)
```

```python
train = pd.read_csv('sample_data/california_housing_train.csv')
train['housing_median_age_bin'] = pd.cut(train['housing_median_age'], 3, labels=["new", "medium", "old"])
train
```

```python
a = train['housing_median_age_bin']
le = LabelEncoder()
le.fit(a)
le.transform(a) # new = 1, medium = 0, old = 2
```

#### 2. 원-핫인코더 사용하기1

```python
"""
범주형 데이터(카테고리 데이터)를 1과 0으로 이루어진 벡터로 변환하는 데 사용됩니다. 
범주형 데이터를 머신러닝 모델에 사용할 수 있도록 수치형으로 변환하는 데 널리 사용됩니다.
"""
from sklearn.preprocessing import OneHotEncoder
```

```python
train = pd.read_csv('sample_data/california_housing_train.csv')
train['housing_median_age_bin'] = pd.cut(train['housing_median_age'], 3, labels=["new", "medium", "old"])
train
```

```python
"""
new -> [0., 1., 0.]
medium ->  [1., 0., 0.]
old -> [0., 0., 1.]
"""

a = train[['housing_median_age_bin']]
le = OneHotEncoder(sparse_output=False) # OneHotEncoder 객체 생성
le.fit(a) # 고유한 범주값 학습, 학습된 고유 범주는 ["new", "medium", "old"]
le.transform(a) # a 데이터를 원-핫 인코딩으로 변환
```

```python
housing_median_age_bin_encoding = le.transform(a)
pd.DataFrame(housing_median_age_bin_encoding) # 원-핫 인코딩된 데이터를 새로운 데이터프레임으로 변환
```

```python
new_train = pd.concat([train, pd.DataFrame(housing_median_age_bin_encoding)], axis=1) # 열(column)을 기준으로 데이터프레임을 병합(수평 방향으로 연결)
new_train
```

```python
"""
기존 열 이름이 숫자(0, 1, 2)로 되어 있는 경우, 각각 다음과 같은 새로운 이름으로 변경합니다:
0 → housing_median_age_bin_new
1 → housing_median_age_bin_old
2 → housing_median_age_bin_medium
"""

new_train = new_train.rename(columns={0: 'housing_median_age_bin_new', 1: 'housing_median_age_bin_old', 2: 'housing_median_age_bin_medium'})
new_train
```

#### 3. 원-핫인코더 사용하기2

```python
train['housing_median_age_bin'] = pd.cut(train['housing_median_age'], 3, labels=["new", "medium", "old"])
train
```

```python
# pd.get_dummies: 범주형 데이터를 원-핫 인코딩하여 새로운 열을 생성하는 함수

"""
housing_median_age_bin 열의 고유값: ['new', 'medium', 'old'].
각 고유값에 대해 새로운 열을 만듭니다: housing_median_age_bin_new, housing_median_age_bin_medium, housing_median_age_bin_old.
0과 1로 값 지정: 원래 데이터가 해당 범주라면 1, 아니면 0으로 표시.
"""

pd.get_dummies(train, columns=['housing_median_age_bin'])
```

####  4. 선형회귀

```python
from sklearn.linear_model import LinearRegression
```

```python
x = [[1], [2], [6], [8]] # 독립변수(입력값)
y = [6, 8, 9, 11] # 종속변수(출력값)

"""
LinearRegression(): Scikit-learn에서 제공하는 선형 회귀 모델.
이 모델은 𝑦=𝑎𝑥+𝑏 형태의 직선 방정식을 학습합니다.
"""
model = LinearRegression()
model.fit(x, y)
print("기울기:", model.coef_)     # 기울기 a
print("절편:", model.intercept_)  # 절편 b

# Y=0.59541985*X + 5.082135523613962
```

```python
a = 1
y = a * 0.59541985 + model.intercept_
y
```
